```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Data
file_path = 'Kelas F_Student AI Usage.csv'
df = pd.read_csv(file_path)

# 2. Pengolahan Data
# Mengelompokkan berdasarkan ai_tools_used dan mencari rata-rata daily_screen_time_hours
data_grafik = df.groupby('ai_tools_used')['daily_screen_time_hours'].mean().sort_values(ascending=False).reset_index()

# 3. Pengaturan Visualisasi
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Membuat Bar Chart
ax = sns.barplot(
    x='ai_tools_used',
    y='daily_screen_time_hours',
    data=data_grafik,
    palette='viridis'
)

# 4. Menambahkan Label Angka di Atas Bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f} jam',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points',
                fontsize=11,
                fontweight='bold')

# 5. Menambahkan Judul dan Keterangan Sumbu
plt.title('Rata-rata Durasi Screen Time Harian Berdasarkan Software AI yang Digunakan', fontsize=14, pad=20)
plt.xlabel('Software AI yang Digunakan', fontsize=12)
plt.ylabel('Rata-rata Screen Time (Jam)', fontsize=12)

plt.tight_layout()
plt.show()
```
