```python
import matplotlib.pyplot as plt
import pandas as pd  # <--- Pastikan pandas di-import
import seaborn as sns

# 1. LOAD DATA KAMU TERLEBIH DAHULU
# Ganti 'nama_file_kamu.csv' sesuai dengan nama file data yang kamu punya
df = pd.read_csv('Kelas F_Student AI Usage.csv') 

# ===========================================================================
# KATEGORI 3: KORELASI (DURASI LAYAR VS WAKTU BELAJAR)
# ===========================================================================
plt.figure(figsize=(8, 5))

# Sekarang Python sudah tahu 'df' itu apa, jadi baris ini tidak akan error lagi
sns.stripplot(
    x='daily_screen_time_hours', 
    y='study_hours_per_day', 
    data=df, 
    color='purple', 
    alpha=0.6,      
    jitter=0.15,    
    size=6
)

# Kustomisasi kelengkapan grafik laporan
plt.title('Korelasi Durasi Layar Harian vs Waktu Belajar (Jittered)', fontsize=12, fontweight='bold')
plt.xlabel('Durasi Layar Harian (Jam)')
plt.ylabel('Waktu Belajar (Jam/Hari)')
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
