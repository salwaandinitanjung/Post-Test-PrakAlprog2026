```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# 1. LOAD DATA
df = pd.read_csv('Kelas F_Student AI Usage.csv')

# 2. DEFINISI KONSTANTA ESTETIKA (Konsisten dengan tema laporan)
ACCENT = '#2C3E50'      # Warna teks utama / aksen gelap
BG_COLOR = '#FFFFFF'    # Warna latar belakang plot
GRID_COLOR = '#BDC3C7'  # Warna garis grid
PLOT_COLOR = '#27AE60'  # Warna hijau premium (Emerald) untuk distribusi

# 3. MEMBUAT CANVAS PLOT
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_facecolor(BG_COLOR)

# 4. MEMBUAT HISTOGRAM + KDE (Dengan penyesuaian estetika premium)
sns.histplot(
    df['grades_before_ai'], 
    kde=True, 
    color=PLOT_COLOR, 
    discrete=True,                 # Memastikan tiap balok mewakili unit nilai bulat yang pas
    alpha=0.45,                    # Mengatur kelembutan warna isi balok histogram
    kde_kws={'bw_adjust': 1.2},    # Menghaluskan lekukan kurva tren agar melandai normal
    line_kws={'linewidth': 2.5},   # Mempertegas garis kurva KDE agar terlihat jelas
    edgecolor='#FFFFFF',           # Memberi batas putih tipis antar balok agar lebih rapi
    ax=ax                          # Plot ke axis yang sudah dibuat
)

# 5. KUSTOMISASI LABEL & TIPOGRAFI
ax.set_title(
    'Kategori 4: Sebaran Distribusi Nilai Awal Siswa (Global)', 
    fontsize=13, 
    fontweight='bold', 
    color=ACCENT,
    pad=15
)
ax.set_xlabel('Nilai Sebelum Mengenal AI', fontsize=11, color=ACCENT, labelpad=8)
ax.set_ylabel('Frekuensi Sensus', fontsize=11, color=ACCENT, labelpad=8)

# 6. MERAPIKAN GRID DAN GARIS TEPI (SPINES)
# Hanya memunculkan grid pada sumbu Y karena ini grafik frekuensi/tinggi balok
ax.grid(axis='y', linestyle='--', alpha=0.4, color=GRID_COLOR)

# Menghilangkan border atas dan kanan agar grafik terlihat lega dan modern
ax.spines[['top', 'right']].set_visible(False)  
ax.spines[['left', 'bottom']].set_color(GRID_COLOR)  # Melembutkan warna border sisa

# 7. MENYESUAIKAN LAYOUT
plt.tight_layout()
plt.show()
```
